```python
with open ('Day2_2021.txt') as file:
    data = file.readlines()
    data = [line.strip() for line in data]
    
```

```python
data
```

```python
forward = []
down = []
up = []

#Listen trennen und letzte Zahl einf√ºgen
for i in data:
    if 'forward' in i: forward.append(int(i[-1:]))
    if 'down' in i: down.append(int(i[-1:]))
    if 'up' in i: up.append(int(i[-1:]))
```

```python
print(sum(forward) * (sum(down) - sum(up)))
```

```python
space = []
name = []
aim = 0
hor = 0
for j in data:
    if 'forward' in j: space.append(int(j[-1:]))
    if 'down' in j: space.append(int(j[-1:]))
    if 'up' in j: space.append(int(j[-1:]))
        
for k in data:
    if 'forward' in k: name.append(k[:1])
    if 'down' in k: name.append(k[:1])
    if 'up' in k: name.append(k[:1])
```

```python
name
```

```python
with open('Day2_2021.txt') as f:
    line = f.readline().split()

    horizontal = 0
    depth = 0
    aim = 0

    while line:
        if line[0] == "forward":
            horizontal += int(line[1])
            depth += int(line[1]) * aim
        if line[0] == "down":
            aim += int(line[1])
        if line[0] == "up":
            aim -= int(line[1])
        line = f.readline().split()

print(horizontal * depth)
```
