```python
def puzzle_einlesen(datei):
    with open(datei) as file:
        return [x for x in file.read().split('\n\n')]

puzzle = puzzle_einlesen('Data_day_6')
print(puzzle)
```

```python
def löse(puzzle):
    return sum([len({buchstabe for buchstabe in group.replace('\n','')}) for group in puzzle])
#Sets wählen weil diese keine doppelten Buchstaben beachten
print(löse(puzzle))
```

```python
from collections import Counter 
#jeder Buchstabe wird gezählt

def löse2(puzzle):
    summe=0
    for group in puzzle:
        gruppengrösse = group.count('\n') +1 #letztes \n fehlt
        antworten = Counter(group)
        summe += sum([1 for anz in antworten.values() if anz == gruppengrösse])
    return summe
    
print(löse2(puzzle))
```
